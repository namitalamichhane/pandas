#Boolean Selection Multiple Conditions
#Different logical operators for boolean Series
- & for and (ampersand character)
- | for or (pipe character)
- ~ for not (tilde character)

```python
#& condition
filt1 = bikes['tripduration'] > 1000
filt2 = bikes['gender'] == 'Male'
filt = filt1 & filt2
bikes[filt].head()
```

```python
#Multiple conditions in one line
# Use parentheses to separate conditions

filt = (bikes['tripduration'] > 1000) & (bikes['gender'] == 'Male')
bikes[filt].head(3)
```

```python
#Using OR condition
```

```python
filt = (bikes['tripduration'] > 1000) | (bikes['gender'] == 'Male')
bikes[filt].head(3)
```

```python
#Inverting a condition with the not operator
filt = bikes['tripduration'] > 1000
filt2 = bikes['gender'] == 'Male'
filt = filt1 | filt2
bikes[~filt].head(3)
```

```python
# The isin method - to test equality of a certain number of strings
#Use this instead of using multiple OR conditions
#eutai category ma vaye milcha
filt = bikes['events'].isin(['rain','snow','tstorms','sleet'])
bikes[filt].head()
```

```python
filt1 = filt = bikes['events'].isin(['rain','snow','tstorms','sleet'])
filt2 = bikes['tripduration']>5000
bikes[filt1 & filt2]
```

```python
filt = (bikes['events'] =='snow') & (bikes['temperature'] > 30)
bikes[filt]. head(3)
```
